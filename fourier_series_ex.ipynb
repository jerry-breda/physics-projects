```python
import matplotlib.pyplot as plt
import numpy as np, math
```

# Problems, Section 3

### For each of the following combinations of a fundamental musical tone and some of its
### overtones, make a computer plot of individual harmonics (all on the same axes) and then
### a plot of the sum.

### **Number 1**

```python
pi = math.pi
t = np.arange(0, 5*pi, 0.1)
a1 = np.sin(t)
a2 = np.sin(3*t)/9
y = a1-a2
```

```python
ax1 = plt.subplot(212)
ax1.plot(t,y,color='red')

ax2 = plt.subplot(221)
ax2.plot(t,a1,color='green')

ax3 = plt.subplot(222)
ax3.plot(t,a2,color='blue')

plt.show()
```

### **Number 2**

```python
b1 = 2*np.cos(t)
b2 = np.cos(2*t)
y2 = b1 + b2
```

```python
ax4 = plt.subplot(212)
ax4.plot(t, y)

ax5 = plt.subplot(221)
ax5.plot(t, b1)

ax6 = plt.subplot(222)
ax6.plot(t, b2, color='black')

plt.show()
```

### **Number 3**

```python
c1 = np.sin(pi*t)
c2 = np.sin(2*pi*t)
c3 = np.sin(3*pi*t)/3
y3 = c1+c2+c3
```

```python
ax7 = plt.subplot(411)
ax7.plot(t, c1)


ax8 = plt.subplot(412,sharex=ax7,sharey=ax7)
ax8.plot(t, c2)

ax9 = plt.subplot(413,sharex=ax7,sharey=ax7)
ax9.plot(t, c3)

ax10 = plt.subplot(414,sharex=ax7,sharey=ax7)
ax10.plot(t, y3,color='black')

plt.show()
```
